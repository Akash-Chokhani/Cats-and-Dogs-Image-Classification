```python
#Extract training data from zip folder
from zipfile import ZipFile

file_path="train.zip"
zip=ZipFile(file_path,'r')
zip.extractall()
```

```python
#Create subdirectories for cats and dogs
from os import listdir, makedirs
from shutil import copyfile

makedirs('data')
makedirs('data/dog')
makedirs('data/cat')

for img in listdir('train'):
  if img.startswith('cat'):
    copyfile('train/'+img, 'data/cat/'+img)
  else:
    copyfile('train/'+img, 'data/dog/'+img)
```

```python
#Create a data generator
from keras.preprocessing.image import ImageDataGenerator

data_gen=ImageDataGenerator(rescale=1/255)
data=data_gen.flow_from_directory('data', target_size=(200,200),class_mode='binary', batch_size=32)
```

```python
#Model
from keras.layers.serialization import activation
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten

model=Sequential([
    Conv2D(8, (3,3), padding='same', activation='relu', kernel_initializer='he_uniform', input_shape=(200,200,3)),
    MaxPooling2D(2, 2),
    Conv2D(16, (3,3), padding='same', activation='relu', kernel_initializer='he_uniform'),
    MaxPooling2D(2, 2),
    Conv2D(32, (3,3), padding='valid', activation='relu', kernel_initializer='he_uniform'),
    MaxPooling2D(2, 2),
    
    Flatten(),
    Dense(1, activation='sigmoid', kernel_initializer='he_uniform')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
model.fit(data, steps_per_epoch=len(data), epochs=10)
```
